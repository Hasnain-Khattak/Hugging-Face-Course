```python
from transformers import pipeline
import torch
pipe = pipeline("translation", model="facebook/nllb-200-distilled-600M", torch_dtype=torch.bfloat16)
```

```python
text = 'Hello my name is Hasnain and I am from Pakistan.'
translate = pipe(text,
                 src_lang='eng_Latn',
                 tgt_lang='spa_Latn')
```
