```python
import transformers
import torch
```

```python
from transformers import pipeline
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
sentences = ['I love you',
             'I hate you',
             'You are my love']
```

```python
embadding1 = model.encode(sentences, convert_to_tensor=True)
```

```python
embadding1
```

```python
sentences2 = ['I love to play cricet',
             'I hate to watch this movie',
             'You you my best friend']
```

```python
embadding2 = model.encode(sentences2, convert_to_tensor=True)
```

```python
embadding2
```

```python
cosin = util.cos_sim(embadding1, embadding2)
```

```python
print(cosin)
```

```python
for i in range(len(sentences)):
    print("{} \t\t {} \t\t Score: {:.4f}".format(sentences[i],
                                                 sentences2[i],
                                                 cosin[i][i]))
```
