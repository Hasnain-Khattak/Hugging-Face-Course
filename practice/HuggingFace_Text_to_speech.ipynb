```python
from transformers import pipeline

pipe = pipeline("text-to-speech", model="kakao-enterprise/vits-ljs")
```

```python
text = 'Hey there, I am Hasnain and I love this movie and the person very much. Thanks'
```

```python
voice = pipe(text)
```

```python
from IPython.display import Audio as IPythonAudio

IPythonAudio(voice["audio"][0],
             rate=voice["sampling_rate"])
```
